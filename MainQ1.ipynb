```python
# Account 클래스 생성
import random
class 계좌생성:
    생성된계좌 = 0
    def __init__(self, 예금주, 용도, 잔액):
        self.예금주 = 예금주
        self.용도 = 용도
        self.잔액 = 잔액
        self.입금횟수 = 0
        self.입출금내역 = []
        self.계좌번호 = 계좌생성.계좌번호생성()
        계좌생성.생성된계좌 += 1
        print(f"은행이름: SC은행, "
              f"용도: {self.용도}, "
              f"고객명: {self.예금주}, "
              f"계좌번호: {self.계좌번호}, "
              f"보유금: {self.잔액:,}원")
    # 계좌번호 생성 메서드
    @staticmethod
    def 계좌번호생성():
        return (''.join(str(random.randint(0, 9)) for _ in range(3)) + "-" +''.join(str(random.randint(0, 9)) for _ in range(2)) + "-" +''.join(str(random.randint(0, 9)) for _ in range(6)))
    # 생성된 계좌 수
    @classmethod
    def 생성된계좌수(cls):
        return cls.생성된계좌
    # 입금 메서드
    def 입금하기(self, 입금액):
        if 입금액 <= 0:
            print("입금은 최소 1원 이상만 가능합니다.")
            return
        self.잔액 += 입금액
        self.입금횟수 += 1
        self.입출금내역.append(("입금", 입금액))
        # 5회마다 이자 지급
        if self.입금횟수 % 5 == 0:
            이자금액 = int(self.잔액 * 0.01)
            self.잔액 += 이자금액
            self.입출금내역.append(("이자", 이자금액))
            print(f"{입금액:,}원이 입금 되었습니다. 이자가 함께 지급되어 현재 고객님의 통장 잔고는 {self.잔액:,}원 입니다.")
        else:
            print(f"{입금액:,}원이 입금 되었습니다. 현재 고객님의 통장 잔고는 {self.잔액:,}원 입니다.")
    # 출금 메서드
    def 출금하기(self, 출금액):
        if 출금액 <= 0:
            print("출금은 최소 1원 이상만 가능합니다.")
            return
        if 출금액 > self.잔액:
            print("한도초과")
            return
        self.잔액 -= 출금액
        self.입출금내역.append(("출금", 출금액))
        print(f"{출금액:,}원이 출금 되었습니다. 현재 고객님의 통장 잔고는 {self.잔액:,}원 입니다.")
    # 계좌 정보 출력
    def __str__(self):
        return (f"은행이름: SC은행\n"
                f"용도: {self.용도}\n"
                f"예금주: {self.예금주}\n"
                f"계좌번호: {self.계좌번호}\n"
                f"잔고: {self.잔액:,}원")
    # 입출금 내역 조회
    def 입출금내역조회(self):
        print("[입출금 내역]")
        for 종류, 금액 in self.입출금내역:
            if 종류 == "입금":
                print(f"'입금 + {금액:,}원")
            elif 종류 == "이자":
                print(f"'이자' + {금액:,}원")
            elif 종류 == "출금":
                print(f"'출금' - {금액:,}원")
        print("   ")
        print(f"현재 잔액: {self.잔액:,}원")

```

```python
H = 계좌생성('허재민', '예금', 0)
J = 계좌생성('허재민', '적금', 0)
M = 계좌생성('허재민', '생활비', 0)
전체계좌 = (H, J, M)
print(계좌생성.생성된계좌수())
```
